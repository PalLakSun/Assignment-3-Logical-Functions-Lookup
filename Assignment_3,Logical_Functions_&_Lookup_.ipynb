=LOOKUP(B162,B155:B158,D155:D158)

VLOOKUP is used because:

It searches vertically down the first column of a table

It finds the row where Employee ID = E103

It returns the value from a specified column in the same row (Department)

This makes VLOOKUP perfect for Employee databases

We can also use  look up and Xlook up also.Xlook up is flexible option  as it   searches the value in either of the ways top to bottom or bottom to top
